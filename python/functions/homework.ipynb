```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def find_owner():
    document_number = input("Введите номер документа: ")
    
    for x in documents:
        if x['number'] == document_number:
            print(x['name'])
            return

    print('Документ не найден в базе')   

def find_loc():
    document_number = input("Введите номер документа: ")
    val = directories.items()
    for k,v in val:
        for name in v:
            if name == document_number:
                print('Документ на полке' + k)
                return
    print('Документ не найден в базе')         
            
    
        
                
                
def all_info():
    for x in documents:
        number = x['number']
        type_ = x['type']
        name = x['name']
        for line in directories:
            for item in directories[line]:
                if item == number:
                    shelf = line
        print(f'№: {number} Тип: {type_} Имя: {name} Номер полки: {shelf}')
        

def add_loc():
    number = input('Введите номер полки')
    if number in directories.keys():
        shelf = ', '.join(list(directories.keys()))
 
        
        print(f'Такая полка уже существует. Текущий перечень полок: {shelf}') 
                          
    else:                 
        directories[number] = []
        print(f"Полка добавлена. Текущий перечень полок: {', '.join(list(directories.keys()))}")

                              
def del_loc():
        number = input('Введите номер полки')
        if len(directories[number])== 0:
            del (directories[number])
                 
            print(f"Полка удалена. Текущий перечень полок: {', '.join(list(directories.keys()))}")
        
        else:
            print(f"На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {', '.join(list(directories.keys()))}")

def main():
    command_type = input("Введите команду: ")
    if command_type == 'p':
        find_owner()
    elif command_type == 's':
        find_loc()
    elif command_type == 'ds':
        del_loc()
    elif command_type == 'l':
        all_info()
    elif command_type == 'as':
        add_loc()
    else:
        print('Не верная команда!')
        

main()
```
